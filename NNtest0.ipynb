```r
library('keras')


fxy <- function(x,y)
{
  (x^2)*(x + 2*y)
  #(x + 2*y)
}

sample_size <- 100

#x <- rpois(n = sample_size, lambda = 1)
#y <- rpois(n = sample_size, lambda = 2)

x <- rnorm(n = sample_size)
y <- rnorm(n = sample_size)


z <- fxy(x,y)

toy4_big_keras <- keras_model_sequential() 


toy4_big_keras %>% 
  layer_dense(units = 20, activation = 'relu') %>% 
  #layer_dropout(rate = 0.4) %>% 
  layer_dense(units = 20, activation = 'relu') %>%
  #layer_dropout(rate = 0.3) %>%
  layer_dense(units = 1, activation = 'linear')

toy4_big_keras %>% 
  compile(loss = "mse",
          optimizer = optimizer_rmsprop())


history <- toy4_big_keras %>% 
              fit(x = cbind(x,y), y = fxy(x,y) )

summary(toy4_big_keras)


plot(history)








```

```r
model_pred <- predict(toy4_big_keras, x = cbind(x,y))
#head(model_pred)
```

```r
plot(z, model_pred)
abline(0,1, col = 'red')
toy4_big_keras$fit
```
